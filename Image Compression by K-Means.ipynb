```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
img=plt.imread("peacock1.jpg")
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
type(img)
```

```python
img.shape
```

```python
img
```

```python
img=img/255
```

```python
img
```

```python
# 3d to 2d by combining w and h
w,h,d=img.shape
img_array=img.reshape(w*h,d)
img_array.shape
```

```python
from sklearn.utils import shuffle
#fitting model on a small sub sample of the complete image
img_array_sample=shuffle(img_array,random_state=1)[:1000]
img_array_sample.size
```

```python
kmeans=KMeans(n_clusters=2,random_state=1)
kmeans.fit(img_array_sample)
```

```python
# get labels all centroid on the complete image
labels=kmeans.predict(img_array)
```

```python
labels
```

```python
labels[0]
```

```python
print(kmeans.cluster_centers_)
```

```python
c=kmeans.cluster_centers_
```

```python
labels.size
```

```python
# recreate original image according to labels and each pixels
def recreate_img(c,labels,w,h,d):
    image = np.zeros((w,h,d))
    labels_idx=0
    # now label ech pixels accrding to the limited labels
    for i in range(w):
        for j in range(h):
            image[i][j]=c[labels[labels_idx]]
            labels_idx=labels_idx+1
            #print(labels_idx)
            #print(labels[label_idx])
    return(image)
```

```python
recreate_img(c,labels,w,h,d)
```

```python
plt.figure(1)
plt.axis('off')
plt.title('original image')
plt.imshow(img)
plt.show()

plt.figure(2)
plt.axis('off')
plt.title('reduced image')
plt.imshow(recreate_img(c,labels,w,h,d))
plt.show()
```
